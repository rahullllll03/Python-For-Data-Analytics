```python
import numpy as np

arr1= np.array([10, 30, 20, 10])
arr2= np.array([30, 20, 10, 20])
print(arr1 + arr2)
print(np.add(arr1, arr2))
```

```python
import numpy as np

arr1= np.array([[30, 20], [10, 30]])
arr2= np.array([[30, 20], [10, 20]])
print(arr1 - arr2)
print(np.subtract(arr1, arr2))
```

```python
import numpy as np

arr1= np.array([[30, 20], [10, 30]])
arr2= np.array([[30, 20], [10, 20]])
print(arr1 * arr2)
print(np.multiply(arr1, arr2))
```

```python
import numpy as np

arr1= np.array([[30, 20], [10, 30]])
arr2= np.array([[30, 20], [10, 20]])
print(arr1 / arr2)
print(np.divide(arr1, arr2))
```

```python
arr1= np.array([3, 18, 69, 1])
arr2= np.array([2])
print(np.power(arr1, arr2))
```
