```python
import matplotlib.pyplot as plt
import pandas as pd

brands = ["Oneplus", "Apple", "Samsung", "Redmi", "Vivo", "Nokia"]
x = [7, 8, 9, 5, 7, 3]
y = ["Green", "Pink", "Orange", "Silver", "Blue", "Brown"]
ex= [0.1, 0, 0.1, 0, 0, 0]
plt.pie(x, labels = brands, colors =y, explode=ex, autopct="%.2f")
plt.show()
```

```python
data = pd.read_excel("Employee_data.xlsx")
df = pd.DataFrame(data)
print(df)
```

```python
plt.pie(df["Age"])
plt.show()
```

```python
grouped_by = df.groupby("Job Title")["Salary"].sum()
print(grouped_by)
plt.pie(grouped_by.values, labels = grouped_by.index, autopct="%.2f")
plt.show()
```
