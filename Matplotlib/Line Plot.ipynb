```python
import matplotlib.pyplot as plt
import pandas as pd

x = ["day1", "day2", "day3", "day4", "day5"]
y = [300, 400, 420, 230, 250]
y1= [290, 350, 400, 400, 380]

plt.plot(x,y, marker = "o", label = "Week1" )
plt.plot(x,y1, marker = "o", label = "Week2")
plt.legend()
plt.show()
```

```python
data = pd.read_excel("sample_employee_data.xlsx.xlsx")
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["Name"], df["Department"], marker = "o")
grouped_by = df.groupby("Name")["Department"].sum()
print(grouped_by)
plt.show()
```

```python
plt.plot(grouped_by.index,grouped_by.values, marker= "o")
```
