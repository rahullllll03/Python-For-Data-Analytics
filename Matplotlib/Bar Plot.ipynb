```python
import matplotlib.pyplot as plt

x= ["Rahul", "Aryan", "Rohan", "Harsh", "Sanika"]
y= [98, 77, 88, 90, 95]

plt.bar(x,y)
plt.xlabel("Name of Cousines")
plt.ylabel("Popularity")   
plt.title("Popularity of Different Parts of Cousines")
plt.savefig("bar.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ipl-matches.csv.csv")
df= pd.DataFrame(data)
grouped_by = pd.groupby("Season")["Venue"].sum()
print(grouped_by)
print(df)
```
