```python
import pandas as pd

data1 = {"Emp Id":["E01", "E02", "E03", "E04", "E05", "E06"],
                   "Nmaes":["Rahul", "Aryan", "Rohan", "Harsh", "Om", "Ravi"],
    "Age":[21, 22, 23, 20, 21, 22]}

data2 = {"Emp Id":["E01", "E02", "E07", "E04", "E08", "E06"],
                   "Salary":[50000, 45000, 55000, 30000, 40000, 35000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print()
print(df2)
```

```python
print(pd.merge(df1, df2, on = "Emp Id"))
```

```python
print(pd.merge(df1, df2, on = "Emp Id", how = "left"))
```

```python
print(pd.merge(df1, df2, on = "Emp Id", how = "right"))
```
