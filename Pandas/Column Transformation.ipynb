```python
import pandas as pd

data = pd.read_excel("Employee_data.xlsx")
# print(data)
data.loc[(data["Experience (Years)"]==3), "Salary Increase"] = "Not Increase"
data.loc[(data["Experience (Years)"]>3), "Salary Increase"] = "Increase by 15%"
data.loc[(data["Experience (Years)"]>10), "Salary Increase"] = "Increase by 30%"
print(data)
```
