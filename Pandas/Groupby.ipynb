```python
import pandas as pd

data = pd.read_excel("Employee_Data_With_First_Last_Name.xlsx")
print(data)
```

```python
gp = data.groupby(["Department","Gender"]).agg({"EmployeeID":"count"})
print(gp)
```

```python
gp1 = data.groupby(["Department","Gender"]).agg({"Salary":"max", "Age":"min"})
print(gp1)
```
