```python
import pandas as pd

dict = {"Fruits":["Mango", "Apples", "Banana", "Orange"],
       "Price":[100, 150, 50, 75],
       "Qantity":[15, 10, 7, 10]}

df1 = pd.DataFrame(dict)
print(df1)
```

```python
df2 = df1.copy()
print(df2)
```

```python
df2.loc[0, "Price"] = 120
df2.loc[1, "Price"] = 175
df2.loc[3, "Price"] = 30
df2.loc[0, "Qantity"] = 12
df2.loc[1, "Qantity"] = 17
df2.loc[3, "Qantity"] = 5

print(df2)
print(df1.compare(df2))
```
